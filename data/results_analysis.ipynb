```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data from the LaTeX table (First Chart)
metrics = ["Hit@1", "Hit@3", "Hit@5", "Hit@10", "MRR"]
polo = [0.176, 0.266, 0.310, 0.351, 0.232]
metapath = [0.189, 0.301, 0.352, 0.432, 0.247]
preprocessing = [0.224, 0.361, 0.427, 0.509, 0.318]
embedding = [0.2435, 0.368, 0.432, 0.526, 0.327]

# Calculate the differences with PoLo
diff_metapath = np.array(metapath) - np.array(polo)
diff_preprocessing = np.array(preprocessing) - np.array(polo)
diff_embedding = np.array(embedding) - np.array(polo)

# Set bar width
bar_width = 0.6

# Calculate positions
x = np.arange(len(metrics))

# Apply Seaborn style
sns.set_theme(style="whitegrid")
sns.set_context("talk")

# Define color palette
colors = {
    "Metapath": sns.color_palette("muted")[0],
    "Pre-processing": sns.color_palette("muted")[1],
    "Embedding": sns.color_palette("muted")[2],
}

# Create stacked horizontal bar chart for differences
plt.figure(figsize=(10, 6))
plt.barh(x, diff_metapath, color=colors["Metapath"], height=bar_width, label="Metapath")
plt.barh(x, diff_preprocessing - diff_metapath, color=colors["Pre-processing"], height=bar_width, label="Pre-processing", left=diff_metapath)
plt.barh(x, diff_embedding - diff_preprocessing, color=colors["Embedding"], height=bar_width, label="Embedding", left=diff_preprocessing)

# Add labels and legend
font_size = 14  # Consistent font size
plt.yticks(x, metrics)
plt.xlabel("Increase Score", labelpad=15, fontsize=font_size + 2, fontweight='bold')
plt.ylabel("Metrics", labelpad=15, fontsize=font_size + 2, fontweight='bold')
plt.legend(title="Component", loc="lower right", fontsize=font_size, title_fontsize=font_size + 1)
plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)

# Tighten layout for clean spacing
plt.tight_layout()

# Thinner grid lines
plt.grid(axis='y', linestyle='--', linewidth=1)
plt.grid(axis='x', linestyle='--', linewidth=1)

# Show the plot
plt.show()
```
